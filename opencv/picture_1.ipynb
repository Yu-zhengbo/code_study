```python
import cv2
import numpy as np
import os

basefile = 'D:/qq/2690486436/FileRecv/train/'
open_file = os.path.join(basefile,'open/')
close_file = os.path.join(basefile,'close/')

def fun(len):
    for i in range(len):
        path = os.path.join(close_file,'%d.jpg'%i)
        img = cv2.imread(path)
        rows,cols = img.shape[:2]
        M = cv2.getRotationMatrix2D((cols/2,rows/2),180,3)    #旋转中心，旋转角度，缩放因子
        dst=cv2.warpAffine(img,M,(2*cols,2*rows))
        cv2.imwrite(path,dst)

a = len(os.listdir(open_file))
b = len(os.listdir(close_file))
print(a,b)
fun(a)
path = os.path.join(open_file,'%d.jpg'%2)
print(path)
cv2.destroyAllWindows()
```

```python
from PIL import Image
import os
 
dir_img = "D:/qq/2690486436/FileRecv/train1/open/"
#待处理图片所在地址
dir_save = "D:/qq/2690486436/FileRecv/train1/open/"
#旋转后保存的地址
 
list_img = os.listdir(dir_img)
 
for img_name in list_img:
    pri_image = Image.open(dir_img+img_name)
    tmppath = dir_save + img_name
    pri_image.rotate(180).save(tmppath)
```

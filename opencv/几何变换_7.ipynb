```python
#扩展缩放
import numpy as np
import cv2

img = cv2.imread('1.png')
res = cv2.resize(img,None,fx=2,fy=2,interpolation=cv2.INTER_CUBIC)   #沿着x和y轴放大两倍

height,width = img.shape[:2]
res = cv2.resize(img,(2*width,2*height),interpolation=cv2.INTER_CUBIC)

while(1):
    cv2.imshow('res',res)
    cv2.imshow('img',img)
    if cv2.waitKey(1)&0xFF == 27:
        break
cv2.destroyAllWindows()
```

```python
#图像平移 
import cv2 
import numpy as np

img = cv2.imread('1.png')
rows,cols = img.shape[:2]
M = np.float32([[1,0,100],[0,1,50]])      #移动（100，50）
dst=cv2.warpAffine(img,M,(cols,rows))

cv2.imshow('img',dst)
cv2.waitKey(0)    
cv2.destroyAllWindows()
```

```python
#图像旋转 
import cv2
import numpy as np

img = cv2.imread('flag.jpg')
rows,cols = img.shape[:2]

M = cv2.getRotationMatrix2D((cols/2,rows/2),45,0.6)    #旋转中心，旋转角度，缩放因子
dst=cv2.warpAffine(img,M,(2*cols,2*rows))

cv2.imshow('img',dst)
cv2.imwrite('flag_1.jpg',dst)
cv2.waitKey(0)    
cv2.destroyAllWindows()
```

```python
#仿射变换
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('1.png')
rows,cols,ch = img.shape
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.subplot(121)
plt.imshow(img)
plt.title('Input')
plt.subplot(122)
plt.imshow(dst)
plt.title('Output')
plt.show()
```

```python
#透视变换
import cv2
import numpy as np
from matplotlib import pyplot as plt
img=cv2.imread('1.png')
rows,cols,ch=img.shape
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
M=cv2.getPerspectiveTransform(pts1,pts2)
dst=cv2.warpPerspective(img,M,(300,300))
plt.subplot(121)
plt.imshow(img)
plt.title('Input')
plt.subplot(122)
plt.imshow(dst)
plt.title('Output')
plt.show()
```

```python
#均值滤波
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('1.png')
kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)
plt.subplot(121)
plt.imshow(img)
plt.title('original')
plt.xticks([])
plt.yticks([])
plt.subplot(122)
plt.imshow(dst)
plt.title('averaging')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
#cv2.blur()
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('1.png')
blur = cv2.blur(img,(5,5))
plt.subplot(121)
plt.imshow(img)
plt.title('original')
plt.xticks([])
plt.yticks([])
plt.subplot(122)
plt.imshow(blur)
plt.title('averaging')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
#高斯模糊
blur = cv2.GaussianBlur(img,(5,5),0)  #0是标准差
plt.imshow(blur)
plt.title('Guass')
plt.show()
```

```python
#中值模糊
median = cv2.medianBlur(img,5)
plt.imshow(median)
plt.title('median')
plt.show()
```

```python
#双边滤波  保留边界
blur = cv2.bilateralFilter(img,9,75,75) #领域直径、高斯函数标准差、灰度值相似性高斯函数标准差
plt.imshow(blur)
plt.title('bilateralFilter')
plt.show()
```
